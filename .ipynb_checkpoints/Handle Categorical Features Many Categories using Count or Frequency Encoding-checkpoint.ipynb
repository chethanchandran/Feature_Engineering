```python
import pandas as pd
import numpy as np

data=pd.read_csv("mercedesbenz.csv",usecols=['X1','X2'])
data.head()
```

```python
#Lets find out the unique values in the labels
for col in data.columns:
    print(col,':',len(data[col].unique()),' labels')
```

```python
# Lets obtain the counts for each one of the labels in the variable X2
#Lets capture this in  a dictionary that we can use to re-map the labels
data.X2.value_counts().to_dict()
```

```python
# And now lets replace each label in X2 by its count
# first we make a dictionary that maps each label to the counts
data_dictionary_map=data.X2.value_counts().to_dict()
```

```python
data.head()
```

```python
# Now we replace X2 labels in the dataset data with its value counts
data.X2=data.X2.map(data_dictionary_map)
data.head()
```
